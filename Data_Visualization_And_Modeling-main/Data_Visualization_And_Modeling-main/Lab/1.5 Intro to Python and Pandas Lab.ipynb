```python
import pandas as pd

df = pd.read_csv('brain_stroke.csv', encoding='ISO-8859-1')

df.head()
```

```python
df['age'].plot(kind='hist')
```

```python
gender_counts = pd.Series({
    "Male": ((df['gender'] == 'Male') & (df['stroke'] == 1)).sum(),
     "Female": ((df['gender'] == 'Female') & (df['stroke'] == 1)).sum()})

gender_counts.plot(kind='pie')
```

```python
df['bmi'].plot(kind='hist')
```

```python
df['avg_glucose_level'].plot(kind='hist')
```

```python
smokingCounts = pd.Series({
    "Smoker": ((df['smoking_status'] == 'formerly smoked') | (df['smoking_status'] == 'smokes')).sum(),
     "NonSmoker": ((df['smoking_status'] == 'never smoked') & (df['stroke'] == 1)).sum()})

smokingCounts.plot(kind='pie')
```
